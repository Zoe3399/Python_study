ex ) 1 ≤ N, X ≤ 10,000 부분 이외에는 “1 ~ 10,000 사이 숫자를 입력해주세요” 와 같이 오류 출력이 나올 수 있도록 작성하기

```python
#X보다 작은 수 https://www.acmicpc.net/problem/10871

#N과 X 입력
N, X = map(int, input().split())
#리스트 입력받기
lst = list(map(int, input().split()))
#반복문으로 X보다 작은 수만 출력
print(*(i for i in lst if i < X ))
```

```python
#예외처리 적용
try:
    #N과 X 입력
    N, X = map(int, input().split())
    #조건문과 raise 함수를 이용 예외 발생
    if 1 >= N or X >= 10000 : raise Exception("1~10000사이 숫자를 입력해주세요.")
    #리스트 입력받기
    lst = list(map(int, input().split()))
    #반복문으로 X보다 작은 수만 출력
    print(*(i for i in lst if i < X ))
except Exception as e:
    print(e)
```

문제를 풀면서 코드가 많이 길어진다면, 더 간단한 코드 구현이 가능한지와 함수가 존재 하는지 생각하면서 작성해보세요~!

```python
#입력받기
A, B = int(input()), int(input())
#인덱싱을 사용해서 자릿수 곱해주기
for i in range(1, len(str(B))+1):
    print(A* int(str(B)[-i]))
print(A*B)
```
